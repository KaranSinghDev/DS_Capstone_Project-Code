```python
import requests
import pandas as pd
from graphviz import Digraph
from IPython.display import Image

# Step 1: Fetch Data from SpaceX API (using the REST API)
url = "https://api.spacexdata.com/v4/launches"
response = requests.get(url)
data = response.json()

# Step 2: Convert JSON to DataFrame
df = pd.json_normalize(data)

# Step 3: Basic Data Wrangling (example of cleaning, you can add more preprocessing as needed)
df = df.dropna()  # Drop rows with missing values (simple example)

# Step 4: Show the first few rows of the dataframe to ensure data is correct
print(df.head())

# Step 5: Create the flowchart showing the data collection process using Graphviz
dot = Digraph()

# Add nodes representing each step in the data collection process
dot.node('A', 'Start')
dot.node('B', 'Make SpaceX API GET Request\nEndpoint: /launches')
dot.node('C', 'Receive Data in JSON Format')
dot.node('D', 'Convert JSON to DataFrame')
dot.node('E', 'Data Cleaning and Preprocessing\n(e.g., Drop NA)')
dot.node('F', 'Store Data for Analysis')
dot.node('G', 'End')

# Add edges connecting the steps
dot.edge('A', 'B')
dot.edge('B', 'C')
dot.edge('C', 'D')
dot.edge('D', 'E')
dot.edge('E', 'F')
dot.edge('F', 'G')

# Render the flowchart as a PNG image
dot.render('data_collection_flowchart', format='png', cleanup=True)

# Display the generated flowchart image
Image(filename='data_collection_flowchart.png')
```

```python
#26
import requests
import pandas as pd

# Step 1: Fetch data from SpaceX API
url = "https://api.spacexdata.com/v4/launches"
response = requests.get(url)
data = response.json()

# Step 2: Convert the JSON data into a pandas DataFrame
df = pd.json_normalize(data)

# Step 3: Extract relevant payload data
# Fetch payload information for NASA missions
payload_url = "https://api.spacexdata.com/v4/payloads"
payload_response = requests.get(payload_url)
payload_data = pd.json_normalize(payload_response.json())

# Step 4: Merge payload data with launches data
df = df.explode('payloads')
df = df.merge(payload_data, left_on='payloads', right_on='id', suffixes=('_launch', '_payload'))

# Step 5: Filter for NASA payloads
nasa_payloads = df[df['customers'].apply(lambda x: 'NASA' in x if isinstance(x, list) else False)]

# Step 6: Calculate total payload mass for NASA
total_payload_nasa = nasa_payloads['mass_kg'].sum()

# Step 7: Display the result
print(f"Total Payload Mass Carried by NASA Boosters: {total_payload_nasa} kg")
```

```python
#27
import requests
import pandas as pd

# Step 1: Fetch data from SpaceX API
url = "https://api.spacexdata.com/v4/launches"
response = requests.get(url)
data = response.json()

# Step 2: Convert the JSON data into a pandas DataFrame
df = pd.json_normalize(data)

# Step 3: Fetch payload data from the API
payload_url = "https://api.spacexdata.com/v4/payloads"
payload_response = requests.get(payload_url)
payload_data = pd.json_normalize(payload_response.json())

# Step 4: Merge payload data with launches data
df = df.explode('payloads')
df = df.merge(payload_data, left_on='payloads', right_on='id', suffixes=('_launch', '_payload'))

# Step 5: Filter for Booster Version F9 v1.1
f9_v1_1_payloads = df[df['rocket'] == 'F9 v1.1']  # Assuming 'rocket' contains booster version

# Step 6: Calculate the average payload mass
average_payload_f9_v1_1 = f9_v1_1_payloads['mass_kg'].mean()

# Step 7: Display the result
print(f"Average Payload Mass Carried by Booster Version F9 v1.1: {average_payload_f9_v1_1:.2f} kg")
```

```python
#28
import requests
import pandas as pd

# Step 1: Fetch data from SpaceX API
url = "https://api.spacexdata.com/v4/launches"
response = requests.get(url)
data = response.json()

# Step 2: Convert the JSON data into a pandas DataFrame
df = pd.json_normalize(data)

# Step 3: Fetch payload data from the API
payload_url = "https://api.spacexdata.com/v4/payloads"
payload_response = requests.get(payload_url)
payload_data = pd.json_normalize(payload_response.json())

# Step 4: Merge payload data with launches data
df = df.explode('payloads')
df = df.merge(payload_data, left_on='payloads', right_on='id', suffixes=('_launch', '_payload'))

# Step 5: Find the maximum payload mass
max_payload_mass = df['mass_kg'].max()

# Step 6: Filter the boosters that carried the maximum payload mass
boosters_with_max_payload = df[df['mass_kg'] == max_payload_mass]['rocket'].unique()

# Step 7: Display the result
print(f"Boosters that carried the maximum payload mass ({max_payload_mass} kg):")
print(boosters_with_max_payload)
```

```python
#28
import requests
import pandas as pd

# Step 1: Fetch data from SpaceX API
url = "https://api.spacexdata.com/v4/launches"
response = requests.get(url)
data = response.json()

# Step 2: Convert the JSON data into a pandas DataFrame
df = pd.json_normalize(data)

# Step 3: Filter for launches in the year 2015
df['date_utc'] = pd.to_datetime(df['date_utc'], errors='coerce')
df_2015 = df[df['date_utc'].dt.year == 2015]

# Step 4: Filter for failed landings on drone ships
failed_drone_ship_landings = df_2015[
    (df_2015['cores'].apply(lambda x: x[0]['landing_success'] if isinstance(x, list) and len(x) > 0 else None) == False) &
    (df_2015['cores'].apply(lambda x: x[0]['landing_type'] if isinstance(x, list) and len(x) > 0 else None) == 'ASDS')
]

# Step 5: Extract relevant columns (booster version, launch site, and outcome)
failed_drone_ship_landings = failed_drone_ship_landings[['rocket', 'launchpad', 'date_utc']]

# Step 6: Display the result
print("Failed Drone Ship Landings in 2015:")
print(failed_drone_ship_landings)
```
